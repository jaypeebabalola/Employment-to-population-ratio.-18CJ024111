# Employment to Population Ratio

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import statistics
```

```python
af_countries = [    'Algeria',
                    'Angola',
                    'Benin',
                    'Botswana',
                    'Burkina Faso',
                    'Burundi',
                    'Cameroon',
                    'Cape Verde',
                    'Central African Republic',
                    'Chad',
                    'Comoros',
                    "Côte d'Ivoire",
                    'Democratic Republic of the Congo',
                    'Djibouti',
                    'Egypt',
                    'Equatorial Guinea',
                    'Eritrea',
                    'Eswatini',
                    'Ethiopia',
                    'Gabon',
                    'Gambia',
                    'Ghana',
                    'Guinea',
                    'Guinea-Bissau',
                    'Kenya',
                    'Lesotho',
                    'Liberia',
                    'Libyan Arab Jamahiriya',
                    'Madagascar',
                    'Malawi',
                    'Mali',
                    'Mauritania',
                    'Mauritius',
                    'Morocco',
                    'Mozambique',
                    'Namibia',
                    'Niger',
                    'Nigeria',
                    'Rwanda',
                    'Sao Tome and Principe',
                    'Senegal',
                    'Seychelles',
                    'Sierra Leone',
                    'Somalia',
                    'South Africa',
                    'South Sudan',
                    'Sudan',
                    'Tanzania',
                    'Togo',
                    'Tunisia',
                    'Uganda'
                    'Zambia',
                    'Zimbabwe'
        ]

df   = pd.read_csv('Data.csv')
df = df[ (df['Country or Area'].isin(af_countries) )]
df = df.sort_values('Value')
df = df.set_index('Country or Area')
df
```

```python
for i in range(3):
    l = 1998+i*3
    h = l + 3
    fig, axes  = plt.subplots(1,3, figsize=(10,30), sharey=True)
 
    for idx, year in enumerate(range(l,h)):
        df[df['Year'] == year]['Value'].plot.barh(ax=axes[idx])
        axes[idx].set_title(year)
```

## Yearly Observations

```python
mean_vals = []
for year in range(1998,2007):
    mean_vals.append(df[df['Year'] == year]['Value'].mean())
    
plt.plot(range(1998,2007), mean_vals)
plt.title('Mean Employment-to-Population Ratio of African Countries 1998-2006')
plt.xlabel('Year')
plt.ylabel('Ratio')
```

```python
median_vals = []
for year in range(1998,2007):
    median_vals.append(df[df['Year'] == year]['Value'].median())
    
plt.plot(range(1998,2007), median_vals)
plt.title('Median Employment-to-Population Ratio of African Countries 1998-2006')
plt.xlabel('Year')
plt.ylabel('Ratio')
```

```python
var_vals = []
for year in range(1998,2007):
    var_vals.append(df[df['Year'] == year]['Value'].var())
    
plt.plot(range(1998,2007), var_vals)
plt.title('Variance of Employment-to-Population Ratio of African Countries 1998-2006')
plt.xlabel('Year')
plt.ylabel('Ratio')
```

```python
stdev_vals = []
for year in range(1998,2007):
    stdev_vals.append(df[df['Year'] == year]['Value'].var()**0.5)
    
plt.plot(range(1998,2007), stdev_vals)
plt.title('Standard Dev. of Employment-to-Population Ratio of African Countries 1998-2006')
plt.xlabel('Year')
plt.ylabel('Ratio')
```

```python
range_vals = []
for year in range(1998,2007):
    range_vals.append(df[df['Year'] == year]['Value'].max() - df[df['Year'] == year]['Value'].min())

    
plt.title('Range of Employment-to-Population Ratio Values for African Countries 1998-2006')
plt.xlabel('Year')
plt.ylabel('Ratio')
plt.plot(range(1998,2007), range_vals)
```

```python
iqr_vals = []
def get_iqr(sample):
    q3,q1 = np.percentile(sample, [75 ,25])
    return q3 - q1

for year in range(1998,2007):
    iqr_vals.append(get_iqr(df[df['Year'] == year]['Value']))
    
    
plt.title('Inter Quartile Range Values of Employment-to-Population Ratio of African Countries 1998-2006')
plt.xlabel('Year')
plt.ylabel('Ratio')
plt.plot(range(1998,2007), iqr_vals)
```

```python
siqr_vals = []

for year in range(1998,2007):
    siqr_vals.append(get_iqr(df[df['Year'] == year]['Value'])*0.5)
    
    
plt.title('Semi Inter Quartile Range Values of Employment-to-Population Ratio of African Countries 1998-2006')
plt.xlabel('Year')
plt.ylabel('Ratio')
plt.plot(range(1998,2007), siqr_vals)
```
